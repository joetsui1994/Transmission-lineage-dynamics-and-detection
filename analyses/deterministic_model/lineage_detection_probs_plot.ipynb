#### load in modules

```python
import matplotlib.pyplot as plt
import pandas as pd
```

#### read in simulation output

```python
# observation at T=20
simulation_output_T20 = pd.read_csv('./results/detection_probs_T20.csv')

# observation at T=40
simulation_output_T40 = pd.read_csv('./results/detection_probs_T40.csv')
```

#### plot

```python
# plot curve
fig, ax = plt.subplots(figsize=(6.2, 6.22))

# add vertical/horizontal lines
ax.axhline(y=0, color='black', linewidth=0.5, linestyle='-', alpha=0.5)
ax.axhline(y=1, color='black', linewidth=0.5, linestyle='-', alpha=0.5)
ax.axvline(x=0, color='black', linewidth=0.5, linestyle='-', alpha=0.5)
ax.axvline(x=1, color='black', linewidth=0.5, linestyle='-', alpha=0.5)

# add diagonal line
ax.plot([0, 1], [0, 1], color='black', linewidth=1.5, linestyle='--')

# add lower triangle (greyed out)
ax.fill_between([0, 1, 1, 0], [0, 0, 1, 0], 1, color='darkgrey', alpha=0.2)

# get sampling proportions
sampling_props = simulation_output_T20.sampling_prop.values

# observation at T=20
observe_T20_detection_probs_medians = simulation_output_T20['median'].values
ax.plot(sampling_props, observe_T20_detection_probs_medians, color='#5C89EC', linewidth=4)

# observation at T=40
observe_T40_detection_probs_medians = simulation_output_T40['median'].values
ax.plot(sampling_props, observe_T40_detection_probs_medians, color='#EE8131', linewidth=4)

# adjust axis labels font size
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

# add plot border
for spine in plt.gca().spines.values():
    spine.set_linewidth(3)

plt.show()

# export
fig.savefig('./figures/constant_importation_lineage_detection_probs_median.pdf', bbox_inches='tight')
```
